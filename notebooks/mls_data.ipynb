```python
# Dependencies
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
# MLS 2019 Eastern and Western team stats: 
url = 'https://fbref.com/en/comps/22/2798/2019-Major-League-Soccer-Stats'
```

```python
# Read HTML on page: 
tables = pd.read_html(url)
tables
```

```python
# Find out the type your working with:  
type(tables)
```

```python
# Call table index for Eastern MLS: [0] Eastern, [1] Eastern Home/Away stats, [1] Western, [2] Western Home/Away stats:
eastern_df = tables[0]

# Saving raw data from webscrape: 
eastern_df.to_csv('../data/raw/eastern_mls_raw.csv', encoding='utf-8')
```

```python
# Rename columns:
eastern_df.columns = ['Ranking', 'Squad', 'Matches Played', 'Wins', 'Draws', 'Losses', 'Goals For', 'Goals Against', 'Goal Differential', 'Total Points', 'Expected Goals', "Expected Goals Allowed", "Expected Goals Differiential", "Expected Goals Differiential/ 90 (mins)", 'Spectators', 'Top Team Scorer', 'Goalkeeper', 'Notes' ]
eastern_df.head()
```

```python
# Delete column notes:
del eastern_df['Notes']
eastern_df
```

```python
# Call Squad names to easily match up Clubs:
eastern_df['Squad']
```

```python
# Create a Clubs list to match to Squads.
club = ['New York City FC', 'Atlanta United', 'Philadelphia Union', 'Toronto FC', 'DC United', 'New York Red Bulls', 'New England Revolution', 'Chicago Fire', 
                    'Montreal Impact', 'Columbus Crew',  'Orlando City SC', 'FC Cincinnati']
eastern_df["Club"] = club 
eastern_df.head()
```

```python
# Appending unique value of this table 'eastern' for row in column 'Conference' for later use: 
data = []
def neo_column(table, row_value):
    for i in range(len(table)): # table name of current df variable, column name = " ":, row_value = "string" , int, etc...
        data.append(row_value)
neo_column(eastern_df, "Eastern")
eastern_df["Conference"] = data 
```

```python
# View the table with new column: 
eastern_df
eastern_df.to_csv('../data/interim/eastern_mls_int.csv', encoding='utf-8')
```

```python
# Perform same process on western 
western_df = tables[2]
western_df.head()
# Rename columns:
western_df.columns = ['Ranking', 'Squad', 'Matches Played', 'Wins', 'Draws', 'Losses', 'Goals For', 'Goals Against', 'Goal Differential', 'Total Points', 'Expected Goals', "Expected Goals Allowed", "Expected Goals Differiential", "Expected Goals Differiential/ 90 (mins)", 'Spectators', 'Top Team Scorer', 'Goalkeeper', 'Notes' ]
```

```python
# Delete column notes:
del western_df['Notes']
western_df
```

```python
# Create a Clubs list to match to Squads.
club2 = ['LAFC', 'Seattle Sounders FC', 'Real Salt Lake', 'Minnesota United', 'LA Galaxy', 'Portland Timbers', 'FC Dallas', 'San Jose Earthquakes', 
                    'Colorado Rapids', 'Houston Dynamo',  'Sporting Kansas City', 'Vancouver Whitecaps']
western_df["Club"] = club2 
western_df
```

```python
data2 = [] 
def neo_column2(table, row_value):
    for i in range(len(table)): # table name of current df variable, column name = " ":, row_value = "string" , int, etc...
        data2.append(row_value)
neo_column2(western_df, "Western")
western_df["Conference"] = data2 
western_df.to_csv('../data/interim/western_mls_int.csv', encoding='utf-8')
western_df
```

```python
# Combine Eastern and Western MLS: 
# Stack the DataFrames on top of each other
mls_league_df = pd.concat([eastern_df, western_df], axis=0)
mls_league_df
```

```python
team_id = {
    'Atlanta United': 1,'Chicago Fire': 2,'Colorado Rapids': 3,'Columbus Crew': 4,'DC United': 5,'FC Cincinnati': 6,
    'FC Dallas': 7,'Houston Dynamo': 8,'LA Galaxy': 9,'LAFC': 10,'Minnesota United': 11,'Montreal Impact': 12,
    'New England Revolution': 13,'New York City FC': 14,'New York Red Bulls': 15,'Orlando City SC': 16,'Philadelphia Union': 17,'Portland Timbers': 18,
    'Real Salt Lake': 19,'San Jose Earthquakes': 20,'Seattle Sounders FC': 21,'Sporting Kansas City': 22,'Toronto FC': 23,'Vancouver Whitecaps': 24}
mls_league_df['Club_ID'] = mls_league_df['Club'].map(team_id)
mls_league_df = mls_league_df.set_index('Club_ID')
mls_league_df.to_csv('../data/interim/mls_leagues_int.csv', encoding='utf-8')
mls_league_df
```

```python
# Split dataframe into relevant categories for tables: 
league_df = mls_league_df[['Club','Conference', 'Ranking', 'Matches Played', 'Wins', 'Draws', 'Losses', 'Total Points']]
league_df.to_csv('../data/processed/league.csv', encoding='utf-8')
league_df
```
