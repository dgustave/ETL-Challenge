```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
# MLS 2019
url = 'https://fbref.com/en/comps/22/2798/2019-Major-League-Soccer-Stats'
```

```python
tables = pd.read_html(url)
tables
```

```python
#tables 2 = western conference table
western_df = tables[2]
```

```python
#clean it up
western_df.columns = ['Ranking', 'Squad', 'Matches Played', 'Wins', 'Draws', 'Losses', 'Goals For', 'Goals Against', 'Goal Differential', 'Tota1 Points', 'Expected Goals', "Expected Goals Allowed", "Expected Goals Differiential", "Expected Goals Differiential/ 90 (mins)", 'Attendance (# of fans)', 'Top Team Scorer', 'Goalkeeper', 'Notes' ]
western_df.head()
```

```python
#get rid of notes
del western_df['Notes']
western_df
```

```python
#squad = team
club = ['LAFC', 'Seattle Sounders FC', 'Real Salt Lake', 'Minnesota United', 'LA Galaxy', 'Portland Timbers', 'FC Dallas', 'San Jose Earthquakes', 
                    'Colorado Rapids', 'Houston Dynamo',  'Sporting KC', 'Vancouver Whitecaps']
western_df["Club"] = club 
western_df.head()
```

```python
data = []
def neo_column(table, row_value):
    for i in range(len(table)): 
        data.append(row_value)
neo_column(western_df, "Western")
western_df["Conference"] = data
```
